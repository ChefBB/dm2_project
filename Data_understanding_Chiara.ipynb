### ABOD

```python
print(df.dtypes)
```

```python
# X = df[['awardWins', 'numVotes', 'totalVideos', 'totalCredits', 'userReviewsTotal', 'criticReviewsTotal', 'totalImages', 
#         'numRegions', 'castNumber', 'averageRating', 'quotesTotal']]

df_no_missing = df.dropna()

X = df_no_missing.select_dtypes(include=['int64', 'float64']).values
```

```python
#import StandardScaler
from sklearn.preprocessing import StandardScaler
scl = StandardScaler()
X = scl.fit_transform(X)
```

```python
#Outlier detection with ABOD
from pyod.models.abod import ABOD 
```

```python
# Create the ABOD model with numba disabled
abod = ABOD(contamination=0.1)

# Fit the model
abod.fit(X)
```

```python
outliers = abod.predict(X)
np.unique(outliers, return_counts=True)
```

```python
plt.hist(abod.decision_scores_, bins=20)
plt.axvline(np.min(abod.decision_scores_[np.where(outliers==1)]), c='k')
plt.show()
```

```python
# remove outliers

df_no_missing['outliers'] = outliers
df_no_missing = df_no_missing[df_no_missing['outliers'] == 1]

```

```python
# List of variables to plot
variables_to_plot = ['awardWins', 'numVotes', 'totalVideos', 'totalCredits']

# Create a boxplot for each variable
plt.figure(figsize=(15, 10))
for i, var in enumerate(variables_to_plot, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(y=df_no_missing[var].dropna())
    plt.title(f'Boxplot of {var}')
    plt.tight_layout()

plt.show()
```

### Elliptic Envelope

```python
df_no_missing = df.dropna()
```

```python
from sklearn.covariance import EllipticEnvelope
ellenv = EllipticEnvelope(random_state=0)
ellenv.fit(X)
```

```python
# predict returns 1 for an inlier and -1 for an outlier
outliers = ellenv.predict(X)
```

```python
np.unique(outliers, return_counts=True)
```
