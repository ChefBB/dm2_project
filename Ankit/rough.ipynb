```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('imdb_with_outliers.csv')
df
```

# ABOD

```python


plt.figure(figsize=(10, 6))
colors = df['ABOD_Score'].map({1: 'blue', -1: 'red'})  
plt.scatter(df['t-SNE-1'], df['t-SNE-2'], c=colors, alpha=0.5, s=10)

# Titles and labels
plt.title('t-SNE 2D Plot of Data Points')
plt.xlabel('t-SNE-1')
plt.ylabel('t-SNE-2')

# Grid
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

# ISF

```python

plt.figure(figsize=(10, 6))
colors = df['ISF_Score'].map({1: 'blue', -1: 'red'})  
plt.scatter(df['t-SNE-1'], df['t-SNE-2'], c=colors, alpha=0.5, s=10)

# Titles and labels
plt.title('t-SNE 2D Plot of Data Points')
plt.xlabel('t-SNE-1')
plt.ylabel('t-SNE-2')

# Grid
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

# LOF

```python


plt.figure(figsize=(10, 6))
colors = df['LOF_Score'].map({1: "#1B3AED", -1: "#F40C3E"})  
plt.scatter(df['t-SNE-1'], df['t-SNE-2'], c=colors, alpha=0.6, s=10)

# Titles and labels
plt.title('t-SNE 2D Plot of Data Points')
plt.xlabel('t-SNE-1')
plt.ylabel('t-SNE-2')

# Grid
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
